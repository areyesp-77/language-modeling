````python
from datetime import datetime
````

````python
fecha_nacimiento = input('Introduce tu fecha de nacimiento en formato AA-MM-DD: ')
try:
  cumple = datetime.strptime (fecha_nacimiento, '%y-%m-%d').date()
  hoy = datetime.now().date()
  edad = hoy.year - cumple.year #- ((hoy.month, hoy.day) < (cumple.month, cumple.day))
  print(f'Tu edad es {edad} años.')
except ValueError:
  print('Formato de fecha incorrecto. Usa AA-MM-DD')

````

## Datos

* Numéricos: **int, float, complex**
* Secuencias **str, list, tuple**

Cada tipo de dato acepta cierto tipo de operaciones y funciones, por lo tanto, su sintaxis cambia.


```
>>> '3' + '2' vs. 3 + 2

>>> no quiero que vengas vs. 'no quiero que vengas'
```

Si bien, en lingüística se tiende a trabajar con secuencias por la naturaleza de los datos con los que se trabajan, no se pueden excluir el resto de datos ni de operaciones.


##Operaciones sobre secuencias: STR y LIST

* **Indexado** en un STR: permite recorrer una cadena por medio de sus caracteres individuales.
* Los índices se representan con INT entre corchetes [2]

````python
x = 'Este es un ejemplo de una cadena'      #no olvidar que los STR se deben representar con comillas
````

````python
print (x[1])
````

````python
print (x[10])
````

````python
print (len(x))
````

````python
print (x[-4])
````

````python
print (x[:])
````

````python
print (x[3:9])
````

* Al procesar el STR mediante índices permite realizar operaciones sobre la secuencia (texto), tales como añadir o eliminar índices.



````python
y = x
z = 'no'
print (y+z)
````

````python
y = x
z = 'no'
print (y[0:8]+z)
````

* Segmentación: se obtienen secuencias mayores a un caracter.


````python
i = 'coción'
print (i[:3])
print (i[3:])

j = 'c'
newI= i[:3] + j + i[3:]
print (newI)
````

## **LIST**

Estructura de datos que permite almacenar diferentes elementos.
A diferencia del **STR**, las **LIST** sí pueden modificarse.

* Indexado en una LIST: permite recorrer la lista por medio de sus constituyentes, en lugar de por caracteres
* Los elementos de la LIST deben estar entre corchetes, con comillas y separados por una coma



````python
cadena1 = 'Este es mi ejemplo inicial de una cadena'
print(cadena1)
````

````python
lista1 = ['este', 'también', 'es', 'un', 'ejemplo', 'pero', 'de', 'una', 'lista']
````

````python
print (lista1)
````

````python
print (len(lista1))
````

````python
print (lista1[4])
````

````python
print (lista1[-3])
````

````python
lista2 = lista1 +  ['en', 'Python']
print(lista2)
````

* Las listas pueden ser modificadas usando los índices


````python
lista2[9] = 'con'              #aquí reemplazamos el índice 9 con una nueva unidad
print(lista2)
````

````python
lista2[-2] = 'con'              #aquí reemplazamos el índice negativo -2  con una nueva unidad
print(lista2)
````

#Conclusión:

* Los STR son inmutables, las LIST pueden modificarse

* Al ser modificables, las LIST pueden aceptar métodos, tales como sort() o reverse()

````python
lista2.sort()
print(lista2)
````

````python
lista2.reverse()
print(lista2)
````

````python
lista2.append(':orden agramatical')
print(lista2)
````

````python
#en cada ejecución, los datos en la lista se modifican. Como el método APPEND se ejecutó varias veces, en cada ejecución se añadió un nuevo elemento
print (lista2)
````

````python
lista2.pop(1)                    #elimina un elemento por su índice
print(lista2)
````

````python
lista2.remove('con')
print(lista2)
````

````python
print (len(lista2))
````

````python
lista2.insert(8, "con")
print(lista2)
````
